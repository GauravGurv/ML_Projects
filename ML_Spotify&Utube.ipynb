```python
#----------bar Plot---------------
top10_viewed_songs=df.sort_values(by='Views',ascending=False)[:10]
plt.figure(figsize=(12,6))
plt.title("Figures showing top 10 songs with their Artist")
sns.barplot(x='Track',y='Views',hue='Artist',data=top10_viewed_songs,)
plt.xticks(rotation=45)
plt.show()
```

```python
#----------Line Plot---------------
plt.figure(figsize=(12,6))
plt.title('Analysis of different features of the songs')
sns.lineplot(x='Track',y='Energy',data=top10_viewed_songs,label='Energy')
sns.lineplot(x='Track',y='Speechiness',data=top10_viewed_songs,label='Speechiness')
sns.lineplot(x='Track',y='Instrumentalness',data=top10_viewed_songs,label='Instrumentalness')
sns.lineplot(x='Track',y='Liveness',data=top10_viewed_songs,label='Liveness')
plt.show()
```

```python
#-------------Create a ridgeline plot of the danceability variable, grouped by genre-----------------
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
g = sns.FacetGrid(df, row="Album_type", hue="Album_type", aspect=15, height=.5)
g.map(sns.kdeplot, "Danceability", clip_on=False, shade=True, alpha=1, lw=1.5, bw=.2)
g.map(sns.kdeplot, "Danceability", clip_on=False, color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2, clip_on=False)
g.set(xlim=(0, 1))
g.despine(bottom=True, left=True)
plt.show()
```

```python
#-------------Create a crosstab of the genre and key variables-----------------
cross_tab = pd.crosstab(df['Album_type'], df['Key'])

# Create a stacked bar chart of the crosstab data
cross_tab.plot(kind='bar', stacked=True)

# Add x-axis and y-axis labels and a title
plt.xlabel('Album_type')
plt.ylabel('Count')
plt.title('Number of Tracks by Genre and Key')
plt.show()
```

```python
#-------------Group the data by genre and calculate the mean of the duration_ms variable-----------------
grouped_data = df.groupby('Album_type')['Duration_ms'].mean()

# Create a bar chart of the mean duration_ms for each genre
grouped_data.plot(kind='bar')

# Add x-axis and y-axis labels and a title
plt.xlabel('Genre')
plt.ylabel('Average Duration (ms)')
plt.title('Average Track Duration by Genre')
plt.show()
```

```python
#-------------Group the data by genre and calculate the mean of the danceability and energy variables-----------------
grouped_data = df.groupby('Album_type')[['Danceability', 'Energy']].mean()

# Create a polar plot of the mean danceability and energy for each genre
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)
theta = np.linspace(0, 2*np.pi, len(grouped_data.index), endpoint=False)
bars = ax.bar(theta, grouped_data['Danceability'], width=0.35, bottom=0.0)
bars = ax.bar(theta+np.pi/2, grouped_data['Energy'], width=0.35, bottom=0.0)
plt.xticks(theta, grouped_data.index)
plt.title('Mean Danceability and Energy by Genre')
plt.show()
```

```python
#-------------Select a subset of the data for the 'pop' genre-----------------
pop_df = df[df['Album_type'] == 'pop'][['Danceability', 'Energy', 'Valence', 'Tempo']]

# Calculate the mean of the variables for the 'pop' genre
mean_data = pop_df.mean()

# Create a radar chart of the mean variables for the 'pop' genre
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, polar=True)
theta = ['Danceability', 'Energy', 'Valence', 'Tempo']
values = mean_data.values
ax.plot(theta, values)
ax.fill(theta, values, alpha=0.3)
plt.title('Mean Danceability, Energy, Valence, and Tempo for Pop Tracks')
plt.show()
```

```python
#-------------Create a 3D scatter plot of energy, loudness, and danceability-----------------
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Energy'], df['Loudness'], df['Danceability'], s=10)
ax.set_xlabel('Energy')
ax.set_ylabel('Loudness')
ax.set_zlabel('Danceability')
plt.show()
```

```python
#-------------Create data for the surface plot-----------------
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Create a 3D surface plot of the data
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
